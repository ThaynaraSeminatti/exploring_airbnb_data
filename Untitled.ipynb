```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import export_graphviz
import graphviz
from sklearn.tree import DecisionTreeClassifier
```

```python
dados = pd.read_excel(r"C:\Users\Dell\Documents\CINNECTA\Cientista de Dados_airbnb.xlsx",decimal='.')
```

```python
## renomear colunas
renomear_colunas = {'host_is_superhost':'superhost',
                    'cancellation_policy':'politica_cancelameto',
                    'instant_bookable':'bookavel',
                    'host_total_listings_count':'n_vezes_bokoou',
                    'neighbourhood_cleansed':'limpeza_bairro',
                    'property_type':'tipo_propriedade',
                    'room_type':'tipo_quarto',
                    'accommodates':'acomoda_qnts_pessoas',
                    'bathrooms':'banheiros',
                    'bedrooms':'quartos',
                    'beds':'camas',
                    'bed_type':'tpo_cama',
                    'minimum_nights':'min_noites_aluguel',
                    'number_of_reviews':'n_de_reviews',
                    'review_scores_rating':'nota_avaliação',
                    'review_scores_accuracy':'nota_pontualidade',
                    'review_scores_cleanliness':'nota_limpeza',
                    'review_scores_checkin':'nota_checkin',
                    'review_scores_communication':'nota_comunicacao',
                    'review_scores_location':'nota_localizacao',
                    'review_scores_value':'nota_preço',
                    'price':'preco',
                    'review_scores_rating_na':'n_avaliacoes',
                    'review_scores_accuracy_na':'n_avaliacoes_pontualidade',
                    'review_scores_cleanliness_na':'n_avaliacoes_limpeza',
                    'review_scores_checkin_na':'n_avaliacoes_checkin',
                    'review_scores_communication_na':'n_avaliacoes_comunicacao',
                    'review_scores_location_na':'navaliacoes_localidade',
                    'review_scores_value_na':'n_aaliacoes_valor'}
dados.rename(renomear_colunas, axis=1, inplace=True)
dados.head()



####chegar se as variaveis estão na sua devida classe
#dados.dtypes
# banheiros está como objeto e deve ser um float
```

```python
# CRIAR UM CLSSIFICADOR - ALUGADO POUCAS OU MUITAS VEZES 1 - mais de 30 alugueis 0 - menos de 30 alugueis

dados['idc_aluguel_bom'] = np.nan

for i in range(0,len(dados)):
    if dados['n_vezes_bokoou'][i] >= 60:
        dados['idc_aluguel_bom'][i] = 1
    else:
        dados['idc_aluguel_bom'][i] = 0
   
```

```python
dados['dummi_quarto_compartilhado'] = np.nan
dados['dummi_quarto_privado'] = np.nan
dados['dummi_casa_inteira'] = np.nan
#Shered room e Private room

for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Private room':
        dados['dummi_quarto_privado'][i] = 1
    else:
        dados['dummi_quarto_privado'][i] = 0
        
for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Shared room':
        dados['dummi_quarto_compartilhado'][i] = 1
    else:
        dados['dummi_quarto_compartilhado'][i] = 0
        
for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Enteri home/apt':
        dados['dummi_casa_inteira'][i] = 1
    else:
        dados['dummi_casa_inteira'][i] = 0
```

```python
dados['dummi_tipo_propriedade'] = np.nan
dados['dummi_quarto_privado'] = np.nan
dados['dummi_casa_inteira'] = np.nan
#Shered room e Private room

for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Private room':
        dados['dummi_quarto_privado'][i] = 1
    else:
        dados['dummi_quarto_privado'][i] = 0
        
for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Shared room':
        dados['dummi_quarto_compartilhado'][i] = 1
    else:
        dados['dummi_quarto_compartilhado'][i] = 0
        
for i in range(0,len(dados)):
    if dados['tipo_quarto'][i] == 'Enteri home/apt':
        dados['dummi_casa_inteira'][i] = 1
    else:
        dados['dummi_casa_inteira'][i] = 0
```

```python
####criando o modelo 

x = dados[["preco", 'dummi_quarto_privado','dummi_quarto_compartilhado']] ### ultilizando essas features
y = dados["idc_aluguel_bom"]

SEED = 5
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
(treino_x == 1).sum()
(previsoes == 1).sum()
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()
dummy.fit(treino_x, treino_y)
previsoes = dummy.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia do dummy foi %.2f%%" % acuracia)
```

```python

SEED = 5
np.random.seed(SEED)
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                         stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

modelo = SVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)


### foi criado um modelo que preve o aluguel deste imoveis com acertode de 82% a partir dessas 4 fitures acima
```
